```python
# subsets based on overall Condition
below_average_condition=df[df['OverallCond']<5]
average_condition=df[df['OverallCond']==5]
above_average_condition = df[df['OverallCond']>5]
```

```python
# use ggplot style
plt.style.use('ggplot')

# plot one figure
plt.figure(figsize=(18,6), layout='constrained')
plt.subplot(2,3,1)

# Create custom bins so all are on the same scale
bins = range(df["SalePrice"].min(), df["SalePrice"].max(), int(df["SalePrice"].median()) // 20) # Plot three histograms, with reduced opacity (alpha) so 

# the three histograms will overlap
plt.hist( x=above_average_condition['SalePrice'],bins=bins, color='black',alpha=0.8,edgecolor='black')
plt.hist( x=average_condition['SalePrice'],bins=bins,color='green',alpha=0.3,edgecolor='green')
plt.hist( x=below_average_condition['SalePrice'],bins=bins, color='red',alpha=0.5,edgecolor='red')
plt.title('Distribution of saleprice grouped OverallCond')
plt.xlabel('SalePrice')
plt.ylabel('Number of Houses')

# set the legend 
plt.legend(['above_average_condition','average_condition','below_average_condition'])

# set the xticks
xticks = np.arange(50000, 800000, 50000)
plt.xticks(xticks)

plt.xticks(rotation=90)

plt.show()


# Most sales is made from average condition which is equal to 5
# least sales is made from below_average_condition which is below 5
```

```python
# plot one figure
plt.figure(figsize=(15, 8), layout='constrained')

# subplot one shows how above average condition impacts saleprice
plt.subplot(2,3,1)
plt.hist( x=above_average_condition['SalePrice'],bins=bins, color='black',alpha=0.8,edgecolor='black')
plt.ylabel('No. of Houses')
plt.xlabel('SalePrice')
plt.title('Above average condition')

# subplot two shows how average condition impacts saleprice
plt.subplot(2,3,2)
plt.hist( x=average_condition['SalePrice'],bins=bins,color='green',alpha=0.3,edgecolor='green')
plt.ylabel('No. of Houses')
plt.xlabel('SalePrice')
plt.title('Average condition')

# subplot three shows how below average condition impacts saleprice
plt.subplot(2,3,3)
plt.hist( x=below_average_condition['SalePrice'],bins=bins, color='red',alpha=0.5,edgecolor='red')
plt.ylabel('No. of Houses')
plt.xlabel('SalePrice')
plt.title('Below average condition')
plt.show()

# Most saleprice is gotten from houses with average condition
# The amount of saleprice from the houses with below average condition is the least
```

```python
# Quantity_sales subset
Cond_sales = {}
Condition =df[['OverallCond','SalePrice']]
for i in Condition.OverallCond.unique():
    Cond_sales[i] = [df['SalePrice'][v] for v in df[df["OverallCond"] == i].index]

Total_sales ={}
for key, value in Cond_sales.items():
    Total_sales[key] = sum(value)

sale_T = Total_sales.items()
Sorted_sales = sorted(sale_T, key=lambda x: x[0])
x = dict(Sorted_sales).keys()
y = dict(Sorted_sales).values()


fig, ax = plt.subplots()
ax.bar(x,y)
ax.set_ylabel('SalePrice')
ax.set_xlabel('Overall Condition')
# ax.set_xticklabels(np.arange(0,15,1)+1)
ax.set_title('Sales distribution per Overall Condition');

# The highest saleprice comes from houses with overall condition 5 which is $16,000,000
# The lowest saleprice is made from houses with overall condition btw 1 and 3 and 9
```

```python
# finding IQR
percentile25 = df['SalePrice'].quantile(0.25)
percentile75 = df['SalePrice'].quantile(0.75)

# finding the upper and lower limits
upper_limit = percentile75 + 1.5 * percentile75
lower_limit = percentile25 - 1.5 * percentile25

# finding outliers
df[df['SalePrice'] > upper_limit]
df[df['SalePrice'] < lower_limit]

# trimming outliers
new_df= df[(df['SalePrice'] < upper_limit) & (df['SalePrice']>lower_limit)]
new_df.shape

#  Compare the plots after trimming
plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.histplot(df['SalePrice'],kde=True)
plt.subplot(2,2,2)
sns.boxplot(df['SalePrice'])
plt.subplot(2,2,3)
sns.histplot(new_df['SalePrice'])
plt.subplot(2,2,4)
sns.boxplot(new_df['SalePrice'])
plt.show()
```

```python
# finding IQR
percentile25 = df['YrSold'].quantile(0.25)
percentile75 = df['YrSold'].quantile(0.75)

# finding the upper and lower limits
upper_limit = percentile75 + 1.5 * percentile75
lower_limit = percentile25 - 1.5 * percentile25

# finding outliers
df[df['YrSold'] > upper_limit]
df[df['YrSold'] < lower_limit]

# trimming outliers
new_df= df[(df['YrSold'] < upper_limit)&(df['YrSold']>lower_limit)]
new_df.shape

#  Compare the plots after trimming
plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.histplot(df['YrSold'])
plt.subplot(2,2,2)
sns.boxplot(df['YrSold'])
plt.subplot(2,2,3)
sns.histplot(new_df['YrSold'])
plt.subplot(2,2,4)
sns.boxplot(new_df['YrSold'])
plt.show()
```
